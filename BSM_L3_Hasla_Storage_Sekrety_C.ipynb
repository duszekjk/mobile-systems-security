# C07 — Password Hasher z saltem

To zadanie jest kontynuacją `C05` i pracuje na haśle lokalnego konta `vault-c05-01!`.
Tym razem nie chodzi o prostą demonstrację saltu. Masz zbudować działający hasher haseł dla aplikacji mobilnej.

## Po co to robimy
Po `C05` masz już lokalne konto aplikacji. Sam fakt, że przestaniesz trzymać hasło jawnie, nie rozwiązuje jeszcze problemu.
Jeżeli zapiszesz pojedynczy szybki hash, np. zwykłe `SHA-256(password)`, to po wycieku lokalnych danych atakujący nadal może wykonywać bardzo szybkie zgadywanie offline.
Dlatego w tym zadaniu budujesz hasher, który:
- używa saltu, aby dwa identyczne hasła nie dawały identycznego wyniku,
- używa wielu iteracji, aby podnieść koszt pojedynczej próby,
- zapisuje rekord w formacie, który da się później zweryfikować.

To nie jest jeszcze pełna architektura bezpieczeństwa całej aplikacji, ale jest to już realny krok w stronę sensownego przechowywania lokalnego sekretu.

## Dlaczego PBKDF2
W tym zadaniu używasz **PBKDF2WithHmacSHA256**.
To nie jest dziś najlepszy możliwy wybór dla nowych systemów serwerowych, bo współcześnie zwykle preferuje się konstrukcje takie jak Argon2id albo scrypt.
Mimo to PBKDF2 jest dobrym wyborem dydaktycznym i praktycznie osiągalnym na Androidzie, ponieważ:
- jest dostępny z poziomu standardowych bibliotek Javy,
- ma prosty model użycia,
- pozwala pokazać rolę saltu, iteracji i formatu rekordu,
- nie wymaga dodatkowej biblioteki kryptograficznej do podstawowej implementacji.

## Co dokładnie budujesz
Masz zaimplementować trzy elementy:
1. funkcję wyprowadzającą hash z hasła i saltu,
2. funkcję budującą rekord tekstowy do zapisu,
3. funkcję weryfikującą, czy podane hasło pasuje do zapisanego rekordu.

To ważne rozróżnienie.
Hasher nie ma tylko zwracać jakiegoś ciągu znaków. Ma tworzyć komplet danych, które później pozwolą odtworzyć dokładnie ten sam proces weryfikacji.

## Ustalony model rozwiązania
W tym zadaniu używamy dokładnie tych parametrów:
- algorytm: `PBKDF2WithHmacSHA256`
- liczba iteracji: `120000`
- długość klucza wyjściowego: `256` bitów
- sposób zamiany saltu na bajty: `UTF-8`
- zapis saltu i wyniku: `hex`

Te parametry są z góry ustalone, żeby:
- każdy student implementował ten sam wariant,
- wynik dało się sprawdzać automatycznie,
- nie mieszać zagadnienia strojenia parametrów z samą implementacją hashera.

## Format rekordu
Masz zbudować rekord dokładnie w tej postaci:
`pbkdf2_sha256$<iterations>$<saltHex>$<hashHex>`

Ten format jest celowo tekstowy i jawny.
To ważne: ukrywanie nazwy algorytmu lub liczby iteracji nie daje sensownego bezpieczeństwa.
Przeciwnie, poprawny system zwykle przechowuje te informacje jawnie, tak aby można było:
- zweryfikować stare rekordy,
- migrować parametry w przyszłości,
- rozpoznać, jakim wariantem dany rekord został policzony.

Przykład struktury rekordu:
- prefiks algorytmu: `pbkdf2_sha256`
- separator: znak `$`
- iterations: `120000`
- `saltHex`: salt po zamianie na bajty UTF-8 i potem na hex
- `hashHex`: wynik PBKDF2 zapisany jako hex

## Użyte klasy i funkcje
W tym zadaniu pracujesz na kilku ważnych elementach API:

`PBEKeySpec`
To obiekt opisujący dane wejściowe do PBKDF2.
Przekazujesz do niego:
- hasło jako tablicę znaków,
- salt jako tablicę bajtów,
- liczbę iteracji,
- oczekiwaną długość wyniku w bitach.

`SecretKeyFactory.getInstance(ALGORITHM_NAME)`
Ta fabryka tworzy implementację algorytmu wskazanego przez nazwę.
Tutaj prosisz platformę o wariant `PBKDF2WithHmacSHA256`.

`generateSecret(spec).encoded`
To właściwy moment obliczenia wyniku.
Zwrócony wynik to bajty wyprowadzonego materiału, jeszcze nie zapis tekstowy.

`utf8Bytes(...)`
Helper zamienia tekst saltu na bajty UTF-8.
Jest ważny, bo ten sam salt tekstowy musi zawsze dawać ten sam ciąg bajtów.

`toHex(...)` i `fromHex(...)`
To helpery zamieniające dane binarne na postać tekstową i odwrotnie.
Są potrzebne, bo rekord ma być czytelny i stabilny tekstowo.
